Teste 1: Xbox One vs PC

- H₀ (Hipótese Nula): As médias das notas de usuários são iguais.  
- H₁ (Hipótese Alternativa): As médias são diferentes.  
- Nível de significância (α): 0.05


```python
from scipy import stats

alpha = 0.05

xone_scores = df_recent.loc[df_recent['platform'] == 'XOne', 'user_score'].dropna()
pc_scores = df_recent.loc[df_recent['platform'] == 'PC', 'user_score'].dropna()

stat, p_levene = stats.levene(xone_scores, pc_scores)
print(f'Teste de Levene p-value: {p_levene}')

equal_var = p_levene > alpha  # True se variâncias iguais

t_stat, p_val = stats.ttest_ind(xone_scores, pc_scores, equal_var=equal_var)

print('t-stat:', t_stat, 'p-value:', p_val)
if p_val < alpha:
    print("Rejeitamos H0: as médias são diferentes")
else:
    print("Não rejeitamos H0: não há diferença significativa entre as médias")

```

 Teste 2: Action vs Sports

- H₀ (Hipótese Nula): As médias das notas de usuários são iguais.  
- H₁ (Hipótese Alternativa): As médias são diferentes.  
- Nível de significância (α): 0.05


```python
action_scores = df_recent.loc[df_recent['genre'] == 'Action', 'user_score'].dropna()
sports_scores = df_recent.loc[df_recent['genre'] == 'Sports', 'user_score'].dropna()

stat, p_levene = stats.levene(action_scores, sports_scores)
print(f'Teste de Levene p-value: {p_levene}')

equal_var = p_levene > alpha  # True se variâncias iguais

t_stat, p_val = stats.ttest_ind(action_scores, sports_scores, equal_var=equal_var)

print('t-stat:', t_stat, 'p-value:', p_val)
if p_val < alpha:
    print("Rejeitamos H0: as médias são diferentes")
else:
    print("Não rejeitamos H0: não há diferença significativa entre as médias")
```
